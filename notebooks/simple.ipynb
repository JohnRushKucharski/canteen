```python
import math
import matplotlib.pyplot as plt

def exponential(yo, k, t):
    return yo * math.exp(k * t)

ts = [t for t in range(100)]
def ys(yo, k, ts):
    ys = []
    for t in ts:
        if t == 0: 
            ys.append(exponential(yo, k, t))
        else:
            ys.append(exponential(ys[-1], k, t))
    return ys

ys_up1 = ys(100, 0.01, ts)
ys_dn1 = ys(100, -0.01, ts)
ys_0 = ys(100, 0, ts)

print(ys_up1)
print(ys_dn1)

#plt.plot(ts, ys_up1, label='up 0.01')
plt.plot(ts, ys_dn1, label='down 0.01')
plt.plot(ts, ys_0, label='0')
```

```python
import numpy as np

def outer():
    a = []
    def inner(x):
        a.append(x)
        return np.mean(a), x, x + 1
    return inner

fx = outer()
print(fx(1))
print(fx(10))
```
